```python
import random #난수 생성하기 위한 코드

def get_number() : #함수 생성
    return random.randrange(1,46) #난수 생성 범위 설정해서 반환하기

def set_lotto_numbers() : #함수 생성
    lotto = [] #리스트 생성
    while True :
        num = get_number() #난수를 받기 위해 get_number 함수를 사용하여 num에 생성된 숫자 저장
        
        if lotto.count(num) == 0 : #리스트의 num 수
            lotto.append(num) #리스트에 num 추가하기
            
        if len(lotto) >= 6 : #리스트의 길이가 6보다 크거나 같을 경우
            break #반복문 탈출
            
    return lotto #로또 번호가 담긴 리스트 반환

def match_numbers(user_inputs, lotto_numbers) : #로또 번호와 유저의 번호가 일치하는지 탐색하는 함수
    same_numbers = set(user_inputs) & set(lotto_numbers)
    #리스트 형식이던 user_inputs와 lotto_numbers를 set 형식으로 변경 후 둘의 교집합을 찾아 same_numbers에 저장
    return same_numbers #same_numbers 반환

def check_prize(same_numbers) : #번호가 일치하는 개수에 따라 등수 결정
    if len(same_numbers) == 6 : #same_numbers 내의 원소 개수가 6개일 경우
        print("1st prize! same numbers : ", *same_numbers) #집합 내의 원소를 공백으로 구분하여 출력
    elif len(same_numbers) == 5 : #same_numbers 내의 원소 개수가 5개일 경우
        print("2nd prize! same numbers : ", *same_numbers)
    elif len(same_numbers) == 4 : #same_numbers 내의 원소 개수가 4개일 경우
        print("3rd prize! same numbers : ", *same_numbers)
    elif len(same_numbers) == 3 : #same_numbers 내의 원소 개수가 3개일 경우
        print("4th prize! same numbers : ", *same_numbers)
    else : #same_numbers 내의 원소 개수가 3개 미만일 경우
        print("no prize :(")
    
num = 0 #num 변수의 초기값 설정
print("** Start Lotto **") #로또 프로그램 시작하면 첫 줄에 출력되는 문구
print("1) Set the lotto numbers") 
lotto_numbers = set_lotto_numbers() #set_lotto_numbers 함수에서 반환된 로또 번호 리스트를 lotto_numbers에 저장

print("2) Set the user inputs") #1번에서 로또 번호 생성 후 출력되는 문구
user_inputs = [] #유저의 로또 번호를 저장할 리스트 생성
while True : 
    num = int(input("Enter %d-number : "%(len(user_inputs)+1))) #num은 유저가 입력한 숫자가 저장될 변수
    #%d : 유저가 1번부터 6번까지 6개의 번호를 입력하도록 입력할 번호의 순서를 알려주기 위한 것
    if num < 1 or num > 46 : #유저가 입력한 번호가 지정 범위를 벗어날 경우
        print("No valid number") #잘못 입력되었음을 알려주는 문구 출력
        continue #반복문 계속해서 이어감
    if user_inputs.count(num) == 0 : #유저가 입력한 숫자가 리스트에 없을 경우(중복되지 않은 숫자일 경우)
        user_inputs.append(num) #리스트에 추가
    else : #중복된 숫자를 입력했을 경우
        print("Same number") #중복되었음을 알려주는 문구 추가
        continue #반복문 이어감
    if len(user_inputs) == 6 : #리스트의 길이가 6이 되면
        break #반복문 종료
        
print("3) Match the numbers") 
same_numbers = match_numbers(user_inputs, lotto_numbers) #match_numbers 함수 사용
#user_inputs와 lotto_numbers를 비교하여 둘의 교집합을 same_numbers에 저장

print("4) Check the prize")
check_prize(same_numbers) #일치하는 숫자가 담긴 same_numbers를 check_prize 함수에 넣어 등수를 출력하도록 함

print("5) Sort the lists") #각각 리스트 내의 원소 정렬하기
lotto_numbers.sort()
user_inputs.sort()

print("=== Summary ===")
print("Lotto numbers : ", *lotto_numbers) #lotto numbers의 모든 원소를 공백으로 구분하여 출력
print("User numbers ", *user_inputs) #user_inputs의 모든 원소를 공백으로 구분하여 출력
```
